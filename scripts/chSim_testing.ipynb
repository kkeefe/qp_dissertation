```python
import matplotlib.pyplot as plt
import numpy as np

Ro = 4000
Co = 4000

def aw(R1,R=Ro,C=Co):
    R2 = R - R1
    return 1 / (2 * C * R**2) * (2*R1**2 + R**2 - 2*R*R1 + C*R + 2*R2)

x = np.linspace(1, Ro, Ro*100)
vals = [aw(xi) for xi in x]

plt.plot(x, vals)
print("min R1:", x[np.argmin(vals)])
```

### uniform dists

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# normal dist
u = np.random.uniform(low=0.0, high=1.0, size=int(1e8))
du = [u[-i] - u[i] for i in range(int(len(u)/2))]
plt.hist(du, density=True)


xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
mean, sigma = norm.fit(du)
gaus = norm.pdf(x, mean, sigma)
plt.plot(x, gaus)
print(f"sigma: {sigma}")
```
