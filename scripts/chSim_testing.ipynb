```python
import matplotlib.pyplot as plt
import numpy as np

Ro = 40
Co = 40

def aw(R1,R=Ro,C=Co):
    R2 = R - R1
    return 1 / (2 * C * R**2) * (2*R1**2 + R**2 - 2*R*R1 + C*R + 2*R2)

x = np.linspace(1, Ro, Ro*100)
vals = [aw(xi) for xi in x]

plt.plot(x, vals)
print("min R1:", x[np.argmin(vals)])
```
